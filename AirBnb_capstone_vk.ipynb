#####Q6 - Chart9: Top 10 popular/demanded host based on the reviews or availability_365?

````python
# Q. which are the unique neighbourhood_groups?
# neighbourhood_group : location
# neighbourhood : (area) is a particular part of a town and the people who live there.
popular_host = airbnb_df.groupby(['host_name','room_type']).agg({'number_of_reviews':'sum',
                                                                 'availability_365':'max',}).sort_values('number_of_reviews',
                                                                                            ascending = False).head(10)

top_10_host = popular_host.rename(columns = {'number_of_reviews':'total_review','price':'mean_price','availability_365':'availability_365'}).reset_index()

top_10_host
````
